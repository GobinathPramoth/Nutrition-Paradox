```python
#  Select only particular columns
df_obesity = df_obesity.rename(columns={'Dim1':'Gender','Low':'LowerBound','High' :'UpperBound','NumericValue':'Mean_Estimate','ParentLocation':'Region','TimeDim':'Year','SpatialDim':'Country'})

# changing Gender Data
df_obesity['Gender'] = df_obesity['Gender'].replace({'SEX_BTSX':'Both','SEX_MLE':'Male','SEX_FMLE':'Female'})

# Special code function
def get_country_name(country_code):
    # Dictionary to map special country codes
    special_codes = {
        'WB_UMI': 'Upper-middle-income economies',
        'WB_LI': 'Low-income economies',
        'EUR': 'Europe',
        'SEAR': 'South-East Asia Region',
        'EMR': 'Eastern Mediterranean Region',
        'WB_LMI': 'Lower-middle-income economies',
        'WB_HI': 'High-income economies',
        'AFR': 'Africa Region',
        'AMR': 'Americas Region',
        'GLOBAL': 'Global'
    }

    try:
        if country_code in special_codes:
            return special_codes[country_code]
        country = pycountry.countries.get(alpha_3=country_code)
        if country:
            return country.name
        else:
            return country_code  # Return the original code if not found
    except Exception as e:
        print(f"Error processing country code {country_code}: {e}")
        return country_code # Return the original code in case of other errors

df_obesity['Country'] = df_obesity['Country'].apply(get_country_name)

#Country code functiion
def get_country_name(country_code):
    try:
        country = pycountry.countries.get(alpha_3=country_code)
        if country:
            return country.name
        else:
            return country_code  # Return the original code if not found
    except Exception as e:
        print(f"Error processing country code {country_code}: {e}")
        return country_code # Return the original code in case of other errors

df_obesity['Country'] = df_obesity['Country'].apply(get_country_name)

# Ad new columns CI_Width
df_obesity['CI_Width'] = df_obesity['UpperBound'] - df_obesity['LowerBound']

# Add column Obesity_level
def categorize_obesity(mean_estimate):
    if mean_estimate >= 30:
        return 'High'
    elif 25 <= mean_estimate < 30:
        return 'Moderate'
    else:
        return 'Low'

df_obesity['obesity_level'] = df_obesity['Mean_Estimate'].apply(categorize_obesity)
df_obesity = df_obesity[df_obesity['Year'].between(2012,2022)]
df_obesity
```

**SQL queries for combined analysis**

```python
combined_queries = []

# Query 1: Average mean estimate for both obesity and malnutrition per year
combined_queries.append("""
SELECT
    o.Year,
    AVG(o.Mean_Estimate) AS avg_obesity_mean_estimate,
    AVG(m.Mean_Estimate) AS avg_malnutrition_mean_estimate
FROM
    obesity o
JOIN
    malnutrition m ON o.Year = m.Year
GROUP BY
    o.Year
ORDER BY
    o.Year;
""")

# Query 2: Average mean estimate for both obesity and malnutrition per region
combined_queries.append("""
SELECT
    o.Region,
    AVG(o.Mean_Estimate) AS avg_obesity_mean_estimate,
    AVG(m.Mean_Estimate) AS avg_malnutrition_mean_estimate
FROM
    obesity o
JOIN
    malnutrition m ON o.Region = m.Region
GROUP BY
    o.Region;
""")

# Query 3: Difference between average obesity and malnutrition mean estimate per country (top 10)
combined_queries.append("""
SELECT
    o.Country,
    AVG(o.Mean_Estimate) - AVG(m.Mean_Estimate) AS mean_estimate_difference
FROM
    obesity o
JOIN
    malnutrition m ON o.Country = m.Country
GROUP BY
    o.Country
ORDER BY
    mean_estimate_difference DESC
LIMIT 10;
""")

# Query 4: Average CI width for both obesity and malnutrition per gender
combined_queries.append("""
SELECT
    o.Gender,
    AVG(o.CI_Width) AS avg_obesity_ci_width,
    AVG(m.CI_Width) AS avg_malnutrition_ci_width
FROM
    obesity o
JOIN
    malnutrition m ON o.Gender = m.Gender
GROUP BY
    o.Gender;
""")

# Query 5: Sum of mean estimates for obesity and malnutrition per country and year (top 10)
combined_queries.append("""
SELECT
    o.Country,
    o.Year,
    AVG(o.Mean_Estimate) + AVG(m.Mean_Estimate) AS total_mean_estimate
FROM
    obesity o
JOIN
    malnutrition m ON o.Country = m.Country AND o.Year = m.Year
GROUP BY
    o.Country, o.Year
ORDER BY
    total_mean_estimate DESC
LIMIT 10;
""")

for i, query in enumerate(combined_queries):
    print(f"Combined Query {i+1}:")
    print(query)
    print("-" * 20)
```
